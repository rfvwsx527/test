```python
import numpy as np
import matplotlib.pyplot as plt

# 設定 x 軸的範圍和取樣點
x = np.linspace(0, 2*np.pi, 100)

# 產生三個正弦波
y1 = np.sin(x)  # 振幅為 1，頻率為 1
y2 = np.sin(2*x) # 振幅為 1，頻率為 2
y3 = np.sin(3*x) # 振幅為 1，頻率為 3

# 繪製圖表
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='sin(2x)')
plt.plot(x, y3, label='sin(3x)')

# 設定圖表標題和圖例
plt.title('三個正弦波')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()

# 顯示圖表
plt.show()
```
